**1.8.3 Product sold v/s Income share**

```python
# Calculate the number of products sold for each income bin
income_bins = [0, 20000, 40000, 60000, 80000, 100000]
income_labels = ['<20k', '20k-40k', '40k-60k', '60k-80k', '>80k']
df['Income_bin'] = pd.cut(df['Income'], bins=income_bins, labels=income_labels)
income_counts = df['Income_bin'].value_counts()

# Calculate the percentage of share of sale for each income bin
income_shares = (income_counts / total_products) * 100

print("\nIncome Share of Sales:")
print("")
print(income_shares)

print("\nProduct Share of Sales by Income Bin:")
print("")
print(pd.crosstab(df['Product'], df['Income_bin'], normalize='index') * 100)
```

**Observations**

* The KP781 product is more popular among customers with higher incomes.
* The KP281 product is more popular among customers with lower incomes.
* Customers with higher incomes are more likely to purchase the KP781 product, while customers with lower incomes are more likely to purchase the KP281 and kp481 product.
